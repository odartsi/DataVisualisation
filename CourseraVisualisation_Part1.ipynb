```python
import matplotlib.pyplot as plt
import pandas as pd


data=pd.read_csv('Topic_Survey_Assignment.csv',index_col=0)
data
```

```python
data.sort_values(['Very interested'], ascending=False, axis=0, inplace=True)
data
```

```python
#convert the number in percentages total =2233


data['Very interested'] = pd.Series([round((100 * val / 2233),2) for val in data['Very interested']], index = data.index)

data['Somewhat interested'] = pd.Series([round((100 * val / 2233),2) for val in data['Somewhat interested']], index = data.index)
data['Not interested'] = pd.Series([round((100 * val / 2233),2) for val in data['Not interested']], index = data.index)
data
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

colors_list = ['#5cb85c', '#5bc0de', '#d9534f']

ax=data.plot(kind='bar', figsize=(20, 8), width=0.8, color=colors_list)

plt.title("Percentage of Respondents' interest in Data science areas", size=16)
plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')
plt.gca().spines['left'].set_color('none')
plt.yticks([])
plt.legend(fontsize=14)
plt.xticks(fontsize=14)


x_offset = [-0.28, -0.02, 0.25] # fine tune data for label x-position
for row in range(len(data)):
    offset_idx = 0
    for column in data.columns:
        value = data.ix[row][column]
        ax.annotate(
            f"{value:4.0f}%",
            xy=(row+x_offset[offset_idx], value),
            color="black",
            xycoords="data",
            horizontalalignment="center",
            verticalalignment="bottom", 
            fontsize=14
        )
        offset_idx += 1
```
