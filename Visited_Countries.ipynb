```python
%matplotlib inline
import seaborn as sns
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('countries of the world.csv')
```

```python
visited_countries=['Greece',
                   'Macedonia',
                   'Germany',
                   'France',
                   'Switzerland',
                   'Luxembourg',
                   'Belgium',
                   'Netherlands',
                   'Bulgaria',
                   'Hungary', 
                   'Spain', 
                   'United Kingdom', 
                   'Portugal',
                   'Italy',
                   'Qatar',
                   'Nepal',
                   'Ecuador',
                   'Russia',
                   'United States of America',
                   'Monaco',
                   'Malta']
```

```python
# From GeoPandas, our world map data
worldmap = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))

# Creating axes and plotting world map
fig, ax = plt.subplots(figsize=(20, 10))
worldmap.plot(color="grey", ax=ax)

wm=worldmap.loc[worldmap.name.isin(visited_countries)]
wm.plot(color="lightblue", ax=ax)

## Percentage of the world that I visited:
visited_percentage = len(visited_countries)/len(worldmap.name.unique()) *100
plt.plot([], [], ' ', label='Visited '+str(round(visited_percentage,2))+'% of the world')
legend = plt.legend(fontsize=18)
legend.get_frame().set_facecolor('none')
legend.get_frame().set_edgecolor('b')
legend.get_frame().set_linewidth(0.0)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
